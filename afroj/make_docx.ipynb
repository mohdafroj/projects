# Create DOCX from Markdown

This notebook reads `Trending_Skills_Today.md` and creates a Word document file `Trending_Skills_Today.docx` using `python-docx`.

```python
# Section 1: Import Required Libraries
from pathlib import Path

try:
    from docx import Document
except ImportError:
    raise ImportError('python-docx is required. Install with pip install python-docx')

print('Libraries imported successfully.')
```

```python
# Section 2: Create a Document Object
output_docx = Path('Trending_Skills_Today.docx')
doc = Document()
print(f'Initialized document object: {output_docx}')
```

```python
# Section 3: Add Paragraphs and Text
markdown_path = Path('Trending_Skills_Today.md')
if not markdown_path.exists():
    raise FileNotFoundError(f'{markdown_path} not found in workspace')

for line in markdown_path.read_text(encoding='utf-8').splitlines():
    if line.strip() == '':
        doc.add_paragraph('')
    else:
        doc.add_paragraph(line)

print('Added paragraphs from markdown.')
```

```python
# Section 4: Add Tables
# Add a small example table summarizing sections from the resume.
headers = ['Category', 'Details']
rows = [
    ['Trending focus', 'AI, Cloud-native, Full Stack JavaScript, DevOps, Security'],
    ['Resume profile', 'Full Stack Developer with Python, JS/TS, Vue, React, FastAPI, Nest.js, Laravel'],
]

table = doc.add_table(rows=1, cols=len(headers))
for idx, header in enumerate(headers):
    table.rows[0].cells[idx].text = header

for row in rows:
    cells = table.add_row().cells
    cells[0].text = row[0]
    cells[1].text = row[1]

print('Added a summary table.')
```

```python
# Section 5: Format Text with Styles
# Apply bold formatting to the first paragraph and a heading style.
if doc.paragraphs:
    heading = doc.add_paragraph('Trending Skills Summary')
    heading.style = 'Heading 1'
    heading.add_run('')

paragraphs = doc.paragraphs
if paragraphs:
    para = paragraphs[0]
    run = para.runs[0] if para.runs else para.add_run('')
    run.bold = True
    run.italic = False

print('Applied formatting styles.')
```

```python
# Section 6: Save the Document
doc.save(output_docx)
print(f'Saved document at {output_docx.resolve()}')
```
